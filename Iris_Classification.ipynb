```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
#Encoding
categ_features=['Iris Label']
le=LabelEncoder()
for c in categ_features:
    le.fit(data[c])
    data[c]=le.transform(data[c])
    mapping=dict(zip(le.classes_,le.transform(le.classes_)))
    print(c,':',mapping)
```

```python
x=data.drop('Iris Label',axis=1)
y=data['Iris Label']
```

```python
#5. Splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler, MinMaxScaler
ss=StandardScaler()
mms=MinMaxScaler()
x['sepal length']=ss.fit_transform(x[['sepal length']])
x['sepal width']=ss.fit_transform(x[['sepal width']])
x['petal length']=mms.fit_transform(x[['petal length']])
x['petal width']=mms.fit_transform(x[['petal width']])
```

```python
#Feature Selection
from sklearn.feature_selection import SelectKBest,f_classif
skb=SelectKBest(f_classif,k=3)
skb.fit(x_train,y_train)
```

```python
x_train.columns[skb.get_support()]
```

```python
x_train_sel=x_train[x_train.columns[skb.get_support()]]
x_test_sel=x_test[x_test.columns[skb.get_support()]]
```

```python
#Modelling
#Model 1 : Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(class_weight='balanced')
dt.fit(x_train_sel,y_train)
res_dt=dt.predict(x_test_sel)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
score=accuracy_score(y_test,res_dt)
score
```

```python
plt.figure(figsize=(6,6))
cm=confusion_matrix(y_test,res_dt)
sns.heatmap(cm,annot=True)
```

```python
#Model 2 : Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(class_weight='balanced')
rf.fit(x_train_sel,y_train)
res_rf=rf.predict(x_test_sel)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
score=accuracy_score(y_test,res_rf)
score
```

```python
plt.figure(figsize=(6,6))
cm=confusion_matrix(y_test,res_rf)
sns.heatmap(cm,annot=True)
```

```python
#Model 3 : SVM
from sklearn.svm import SVC
svc=SVC(class_weight='balanced')
svc.fit(x_train_sel,y_train)
res_svc=svc.predict(x_test_sel)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
score=accuracy_score(y_test,res_svc)
score
```

```python
plt.figure(figsize=(6,6))
cm=confusion_matrix(y_test,res_svc)
sns.heatmap(cm,annot=True)
```
